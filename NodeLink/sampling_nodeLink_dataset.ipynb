```python
# Open the file in read mode
file_path = 'triplets.txt'  # Replace 'your_file.txt' with the path to your file
with open(file_path, 'r') as file:
    # Read the content of the file
    content = file.read()

    # Split the content into a list of strings based on whitespace (assuming one number per line)
    numbers_as_strings = content.split()

    # Convert the list of strings to a list of integers
    numbers_as_integers = list(map(int, numbers_as_strings))

# Now, numbers_as_integers contains the list of integers from the file
print(numbers_as_integers)
```

```python
nums = numbers_as_integers
len(nums)
```

```python
len(nums)/3, len(nums)//3
```

```python
matrix = []

for i in range(len(nums)//3):
    matrix.append([nums[3*i], nums[3*i+1], nums[3*i+2]])
matrix
```

# weight thresholding

```python
weight_map = {}

for tri in matrix:
    u = tri[0]
    v = tri[1]
    w = tri[2]
    if u not in weight_map:
        weight_map[u] = w
    else:
        weight_map[u] += w
    
    if v not in weight_map:
        weight_map[v] = w
    else:
        weight_map[v] += w
    
weight_map
```

```python
len(weight_map)
```

## Getting to 100 people

```python
n =64

# Sort the dictionary by values in descending order
sorted_map = dict(sorted(weight_map.items(), key=lambda item: item[1], reverse=True))

# Take the first 100 keys
top_keys = list(sorted_map.keys())[:n]

# Print or use the top 100 keys
print(top_keys)
```

# Making an asoiaf file

```python
final_edges = []

for tri in matrix:
    u = tri[0]
    v = tri[1]
    w = tri[2]
    if u in top_keys and v in top_keys:
        final_edges.append(tri)

len(final_edges), final_edges
```

```python
final_string = "% sym positive\n"

for e in final_edges:
    final_string += f"{e[0]}\t{e[1]}\t{e[2]}\n"

print(final_string)
```

```python
# Your string

# Specify the file path
file_path = "temp.asoiaf"

# Open the file in write mode ('w')
with open(file_path, 'w') as file:
    # Write the string to the file
    file.write(final_string)
```
